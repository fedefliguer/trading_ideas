## 4. Descargo recomendaciones y traduzco

```python
recom, error = ek.get_data(instruments="0#.SPX", fields=['TR.BrkRecLabelEstDate', # Fecha
                                                         'TR.BrkRecLabelEstConfirmDate', # Fecha confirmacion
                                                         'TR.BrkRecLabel', # Recomendación (compra, venta, hold, neutral)
                                                         'TR.TPEstValue', # Target
                                                         'TR.BrkRecLabelEstBrokerName'] # Broker name
                                                        , 
	parameters={'Curn':'USD','SDate':start_date,'EDate':end_date,'Frq':'D'})

recom = recom.sort_values(["Instrument", "Activation Date"], ascending = (True,False))

grouped_recom = recom.groupby("Instrument")
grouped_recom_first_values = grouped_recom.first()
grouped_recom_first_values = grouped_recom_first_values.reset_index()

eng_compra = ['1','1 OUTPERFORM','1 OVERWEIGHT','ACCUMULATE','ADD','ATTRACTIVE','BUY','BUY (1)','BUY/ATTRACTIVE','BUY/CAUTIOUS','MARKET OUTPERFORM','OUTPERFORM','OVERWEIGHT','OVERWT/ATTRACTIVE','OVERWT/CAUTIOUS','OVERWT/IN-LINE','POSITIVE','SECTOR OUTPERFORM','STRONG BUY','TOP PICK']
eng_mantener = ['2','2 MARKET PERFORM','3','3 EQUAL-WEIGHT','BUY/NEUTRAL','CAUTIOUS','EQUALWEIGHT','EQUALWT/ATTRACTIVE','EQUALWT/CAUTIOUS','EQUALWT/IN-LINE','FAIRLY VALUED','HOLD','IN LINE','IN-LINE','LONG','MARKET PERFORM','MARKETPERFORM','NEUTRAL','NEUTRAL (2)','NEUTRAL/ATTRACTIVE','NEUTRAL/CAUTIOUS','NEUTRAL/NEUTRAL','NULL','PEER PERFORM','PEERPERFORM','PERFORM','SECTOR PERFORM','SECTOR WEIGHT']
eng_venta = ['NEGATIVE','REDUCE','SELL','SELL/ATTRACTIVE','SELL/CAUTIOUS','SELL/NEUTRAL','STRONG SELL','TRIM','UNATTRACTIVE','UNDERPERFORM','UNDERWEIGHT','UNDERWT/ATTRACTIVE','UNDERWT/CAUTIOUS','UNDERWT/IN-LINE']
eng_cobertura_susp = ['SUSPENDED COVERAGE']
eng_en_revision = ['UNDER REVIEW']
eng_calificacion_susp = ['RATING SUSPENDED']
eng_no = ['NO RATING SYSTEM','RESTRICTED']

def traduccion(df):
    if df['Broker Rec Descr. - Broker Estimate'] in eng_compra:
        return 'Compra'
    elif df['Broker Rec Descr. - Broker Estimate'] in eng_mantener:
        return 'Mantener'
    elif df['Broker Rec Descr. - Broker Estimate'] in eng_venta:
        return 'Venta'
    elif df['Broker Rec Descr. - Broker Estimate'] in eng_en_revision:
        return 'En revisión'
    elif df['Broker Rec Descr. - Broker Estimate'] in eng_no:
        return '-'
    elif df['Broker Rec Descr. - Broker Estimate'] in eng_calificacion_susp:
        return 'Calificación Suspendida'
    elif df['Broker Rec Descr. - Broker Estimate'] in eng_cobertura_susp:
        return 'Covertura Susp.'

grouped_recom_first_values['recomendacion'] = grouped_recom_first_values.apply(traduccion, axis = 1)  
```

## 5. Descargo último cierre para calcular puntaje de precio actual vs recomendaciones

```python
 # 2.1  Last price

import datetime as DT

hoy = DT.datetime.strptime(fecha, '%Y-%m-%d')
week_ago = hoy - DT.timedelta(days=7)
formato2 = "%Y-%m-%d"
fecha_hoy = hoy.strftime(formato2) 
fecha_7d = week_ago.strftime(formato2) 

start_date=fecha_7d
end_date=fecha_hoy

close, error = ek.get_data(instruments="0#.SPX", fields=['TR.GICSSector','TR.ClosePrice.Date','TR.ClosePrice.Value'], 
    parameters={'Curn':'USD','SDate':start_date,'EDate':end_date,'Frq':'D'})

close=close[close['Close Price'].notnull()]
close = close.sort_values(["Instrument", "Date"], ascending = (True,False))

#Me quedo con el valor más reciente por instrumento
grouped_close = close.groupby("Instrument")
grouped_close_first_values = grouped_close.first()
grouped_close_first_values = grouped_close_first_values.reset_index()

# Comparo precio con target
ultima_recom=pd.merge(grouped_recom_first_values, grouped_close_first_values,on='Instrument', how='left')
ultima_recom['var_target']=ultima_recom['Target Price - Broker Estimate']/ultima_recom['Close Price']-1 # Relación entre el precio target y el actual. Valor positivo es que al precio le falta para llegar al target.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
ultima_recom['var_target_scaled'] = scaler.fit_transform(ultima_recom['var_target'].values.reshape(-1,1)) # Normalizo esa variación

def puntaje(df):
    if df['recomendacion'] =='Venta':
        return 0.5*df['var_target_scaled']
    elif df['recomendacion'] =='Mantener':
        return (0.5*df['var_target_scaled']+0.25)
    elif df['recomendacion'] =='Compra':
        return (0.5*df['var_target_scaled']+0.5)

ultima_recom['puntaje_ultima_recom'] = ultima_recom.apply(puntaje, axis = 1) # Pondero más esas variaciones contra el target si la recomendación es de compra
ultima_recom['puntaje_ultima_recom'] = ultima_recom['puntaje_ultima_recom'].fillna(0)
```

## 8. Junto base, completo nulos en fundamentales, construyo nuevas técnicas con lo que hay

```python
dataset=pd.merge(df_reuters, df_finviz_first_values,on='Ticker', how='left')
dataset=pd.merge(dataset, oe1,on='Ticker', how='left')
dataset=pd.merge(dataset, oe2,on='Ticker', how='left')

dataset1=dataset

dataset1['PEG_topeado'] = np.where((dataset1['Forward P/E/G (Daily Time Series Ratio)'] <0) | (dataset1['Forward P/E/G (Daily Time Series Ratio)']>100 ), 100,dataset1['Forward P/E/G (Daily Time Series Ratio)'])
dataset1['PEG_topeado'] = dataset1['PEG_topeado'].fillna(100)
dataset1['dividend_yield_topeado'] = np.where((dataset1['Dividend yield'] <0) , 0,dataset1['Dividend yield'])
dataset1['dividend_yield_topeado'] = dataset1['dividend_yield_topeado'].fillna(0)
dataset1['PTBV_topeado'] = np.where((dataset1['Price To Book Value Per Share (Daily Time Series Ratio)'] <0) | (dataset1['Price To Book Value Per Share (Daily Time Series Ratio)']>100 ), 100,dataset1['Price To Book Value Per Share (Daily Time Series Ratio)'])
dataset1['PTBV_topeado'] = dataset1['PTBV_topeado'].fillna(100)
dataset1['PTFC_topeado'] = np.where((dataset1['Price To Cash Flow Per Share (Daily Time Series Ratio)'] <0) | (dataset1['Price To Cash Flow Per Share (Daily Time Series Ratio)']>100 ), 100,dataset1['Price To Cash Flow Per Share (Daily Time Series Ratio)'])
dataset1['PTFC_topeado'] = dataset1['PTFC_topeado'].fillna(100)
dataset1['net_margin_topeado'] = np.where((dataset1['Net Profit Margin, (%)'] <0) , 0,dataset1['Net Profit Margin, (%)']/100)
dataset1['net_margin_topeado'] = dataset1['net_margin_topeado'].fillna(0)
dataset1['asset_turnover_topeado'] = np.where((dataset1['Asset Turnover'] <0) , 0,dataset1['Asset Turnover'])
dataset1['asset_turnover_topeado'] = dataset1['asset_turnover_topeado'].fillna(0)

maxim=max(dataset1['Total Debt to Total Equity, Percent'])/100
dataset1['leverage_topeado'] = np.where((dataset1['Total Debt to Total Equity, Percent'] <0) , maxim,dataset1['Total Debt to Total Equity, Percent']/100)
dataset1['leverage_topeado'] = dataset1['leverage_topeado'].fillna(maxim)
dataset1['Earnings Per Share - Mean Growth (This Yr/Last Yr)'] = pd.to_numeric(dataset1['Earnings Per Share - Mean Growth (This Yr/Last Yr)'],errors='coerce')
dataset1['EPS_growth_topeado'] = np.maximum(-1,np.minimum(1,dataset1['Earnings Per Share - Mean Growth (This Yr/Last Yr)']/100))
dataset1['EPS_growth_topeado'] = dataset1['EPS_growth_topeado'].fillna(-1)

#Indicadores técnicos

def flag_medias(df):

    if (df['Moving Average Intersect - 30/60 Day'] == 'nan') & (df['20-day SMA']>df['50-day SMA']):
        return 0.5
    elif (df['Moving Average Intersect - 30/60 Day'] == 'nan') & (df['20-day SMA']<=df['50-day SMA']):
        return -0.5
    if (df['20-day SMA']>df['50-day SMA']) & (df['Moving Average Intersect - 30/60 Day'] == 'P'):
        return 1
    elif (df['20-day SMA']>df['50-day SMA']) & (df['Moving Average Intersect - 30/60 Day'] != 'P'):
        return 0.5
    if (df['20-day SMA']<=df['50-day SMA']) & (df['Moving Average Intersect - 30/60 Day'] == 'N'):
        return -1
    elif (df['20-day SMA']<=df['50-day SMA']) & (df['Moving Average Intersect - 30/60 Day'] != 'N'):
        return -0.5

dataset1['medias_moviles'] = dataset1.apply(flag_medias, axis = 1)  
dataset1['medias_moviles_topeado'] = (dataset1['medias_moviles'] +1)/2

 
def flag_RSI(df):

    if (df['RSI Simple - 14 Days'] <20):
        return 1-(df['RSI Simple - 14 Days']/100)
    elif ((df['RSI Simple - 14 Days'] >=20) and (df['RSI Simple - 14 Days'] <38)):
        return (17/15)-5*(df['RSI Simple - 14 Days']/300)
    elif ((df['RSI Simple - 14 Days'] >=38) & (df['RSI Simple - 14 Days'] <40)):
        return 10-(df['RSI Simple - 14 Days']/4)
    elif ((df['RSI Simple - 14 Days'] >=40) & (df['RSI Simple - 14 Days'] <60)):
        return 0
    elif ((df['RSI Simple - 14 Days'] >=60) & (df['RSI Simple - 14 Days'] <62)):
        return 15-(df['RSI Simple - 14 Days']/4)
    elif ((df['RSI Simple - 14 Days'] >=62) & (df['RSI Simple - 14 Days'] <80)):
        return 8/15-(df['RSI Simple - 14 Days']/60)                  
    elif df['RSI Simple - 14 Days'] >=80:
        return -(df['RSI Simple - 14 Days']/100)

dataset1['RSI'] = dataset1.apply(flag_RSI, axis = 1)  
dataset1['RSI_topeado'] = (1+dataset1['RSI'])/2                                   

# OE
dataset1['OE'] = np.maximum(0,np.minimum(100,100* (dataset1['Close Price']-dataset1['Price - 1 Month Low'])/(dataset1['Price - 1 Month High']-dataset1['Price - 1 Month Low'])))
dataset1['OE_prom'] = (dataset1['OE'] +dataset1['OE1']+dataset1['OE2'])/3

def oe(df):

    if ( (df['OE'] >80) & (df['OE'] > df['OE_prom'])  ) :
        return -0.5
    elif ( (df['OE'] >80) & (df['OE'] <= df['OE_prom']) ) :
        return -1
    if ( (df['OE'] <20) & (df['OE'] > df['OE_prom'])  ) :
        return 1
    elif ( (df['OE'] <20) & (df['OE'] <= df['OE_prom']) ) :
        return 0.5
    if ( (df['OE'] >20) & (df['OE'] <= 80) ) :
        return 0
        
dataset1['OE_calc'] = dataset1.apply(oe, axis = 1)      

dataset1['OE_calc'] = dataset1['OE_calc'].fillna(0)

dataset1['OE_topeado'] = (dataset1['OE_calc'] + 1)/2

#Chartismo

def chartismo(df):

    if ( (df['patron'] =='doublebottom') | (df['patron'] =='headandshouldersinv') | (df['patron'] =='wedgeresistance2') ):
        return 1
    elif ( (df['patron'] =='wedgesupport2') | (df['patron'] =='doubletop') | (df['patron'] =='headandshoulders') ):
        return -1
    elif ( (df['Close Price']>=df['50-day SMA']) & ( (df['patron'] =='wedgedown2') | (df['patron'] =='wedgeresistance') | (df['patron'] =='wedge2') | (df['patron'] =='channelup2') | (df['patron'] =='channeldown2') | (df['patron'] =='channel2') ) ):
        return 1
    elif ( (df['Close Price']>=df['50-day SMA']) & ( (df['patron'] =='wedgedown')  | (df['patron'] =='wedge') | (df['patron'] =='channelup') | (df['patron'] =='channeldown') | (df['patron'] =='channel') ) ):
        return 0.5
    elif ( (df['Close Price']<df['50-day SMA']) & ( (df['patron'] =='wedgeup2') | (df['patron'] =='wedge2') | (df['patron'] =='channelup2') | (df['patron'] =='channeldown2') | (df['patron'] =='channel2') ) ):
        return -1    
    elif ( (df['Close Price']<df['50-day SMA']) & ( (df['patron'] =='wedgeup') | (df['patron'] =='wedge') | (df['patron'] =='channelup') | (df['patron'] =='channeldown') | (df['patron'] =='channel') ) ):
        return -0.5  

dataset1['chartismo'] = dataset1.apply(chartismo, axis = 1)      
dataset1['chartismo_topeado'] = (dataset1['chartismo']+1)/2
dataset1['chartismo_topeado'] = dataset1['chartismo_topeado'].fillna(0)

#Recomendación de analistas externos

dataset1['rendimiento_prom']= np.where(dataset1['Close Price'].astype(float) >0 ,dataset1['Price Target - Mean Estimate'].astype(float)/(dataset1['Close Price'].astype(float))-1,0)
dataset1['rendimiento_max']= np.where(dataset1['Close Price'].astype(float) >0 ,dataset1['Price Target - High'].astype(float)/(dataset1['Close Price'].astype(float))-1,0)
dataset1['rendimiento_min']= np.where(dataset1['Close Price'].astype(float) >0 ,dataset1['Price Target - Low'].astype(float)/(dataset1['Close Price'].astype(float))-1,0)
dataset1['rango_rendimiento']=(dataset1['rendimiento_max']+dataset1['rendimiento_min'])/2
dataset1['total_analistas']= dataset1['Recommendation - Number Of Strong Buy'] + dataset1['Recommendation - Number Of Buy'] + dataset1['Recommendation - Number Of Hold'] + dataset1['Recommendation - Number Of Sell'] + dataset1['Recommendation - Number Of Strong Sell']
dataset1['NPS']= (((dataset1['Recommendation - Number Of Strong Buy'] + dataset1['Recommendation - Number Of Buy'] - dataset1['Recommendation - Number Of Sell'] - dataset1['Recommendation - Number Of Strong Sell'])/(dataset1['Recommendation - Number Of Strong Buy'] + dataset1['Recommendation - Number Of Buy'] + dataset1['Recommendation - Number Of Hold'] + dataset1['Recommendation - Number Of Sell'] + dataset1['Recommendation - Number Of Strong Sell']))+1)/2

#Junto el dataset total con el de ultima recomendacion
ultima_recom1=ultima_recom[["Instrument", "puntaje_ultima_recom"]]
dataset_final=pd.merge(dataset1, ultima_recom1,on='Instrument', how='left')
```

## 10.1. Recomendación Research

```python
#Fundamentals

pond_PEG = 0.2
pond_div_yield = 0.10
pond_PTBV = 0.15
pond_PTFC = 0.15
pond_net_margin = 0.10
pond_asset_turnover = 0.10
pond_leverage = 0.10
pond_EPS_growth = 0.10

dataset_final['fundamentals'] = pond_PEG * dataset_final['PEG_topeado_scaled']+pond_div_yield * dataset_final['dividend_yield_topeado_scaled']+pond_PTBV * dataset_final['PTBV_topeado_scaled']+pond_PTFC * dataset_final['PTFC_topeado_scaled']+pond_net_margin * dataset_final['net_margin_topeado_scaled']+pond_asset_turnover * dataset_final['asset_turnover_topeado_scaled']+pond_leverage * dataset_final['leverage_topeado_scaled']+pond_EPS_growth * dataset_final['EPS_growth_topeado_scaled']

#Técnicos

indicadores = 0.55
chartismo = 0.45

pond_medias_moviles = 0.30
pond_RSI = 0.40
pond_OE = 0.30

dataset_final['tecnicos'] = indicadores * (pond_medias_moviles * dataset1['medias_moviles_topeado'] + pond_RSI * dataset1['RSI_topeado'] + pond_OE * dataset1['OE_topeado']) + chartismo * dataset1['chartismo_topeado']

#Externos

pond_rendimiento_prom = 0.20
pond_total_analistas = 0.20
pond_NPS = 0.40
pond_rango_rendimiento = 0.20

dataset_final['externos'] = 0.50 * (pond_rendimiento_prom * dataset_final['rendimiento_prom_scaled'] + pond_total_analistas * dataset_final['total_analistas_scaled'] + pond_NPS * dataset_final['NPS'] + pond_rango_rendimiento * dataset_final['rango_rendimiento_scaled'] ) + 0.50 * dataset_final['puntaje_ultima_recom']

#Calculo el puntaje total

#Normalizo los subpuntajes
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
dataset_final['fundamentals_scaled'] = scaler.fit_transform(dataset_final['fundamentals'].values.reshape(-1,1))
dataset_final['tecnicos_scaled'] = scaler.fit_transform(dataset_final['tecnicos'].values.reshape(-1,1))
dataset_final['externos_scaled'] = scaler.fit_transform(dataset_final['externos'].values.reshape(-1,1))

#Defino los ponderadores

pond_fundamentals = 0.34
pond_tecnicos = 0.33
pond_externos = 0.33

dataset_final['puntaje_final'] = pond_fundamentals * dataset_final['fundamentals_scaled'] + pond_tecnicos * dataset_final['tecnicos_scaled'] + pond_externos * dataset_final['externos_scaled']


# 8.Recomendación global

recomendacion_global = dataset_final.sort_values(['puntaje_final'], ascending = (False))
recomendacion_global
```
