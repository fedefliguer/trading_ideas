```python
import yfinance as yf
import pandas as pd
import sys
import numpy as np
sys.path.insert(1, 'C:/Users/ffliguer/Documents/algotrading/shared_code')
from iol_algotrading import *
```

```python
tickers = ['BMA.BA','CEPU.BA','CRES.BA','EDN.BA','GGAL.BA','IRSA.BA','LOMA.BA','PAMP.BA','SUPV.BA','TECO2.BA','TGSU2.BA','YPFD.BA']
```

```python
# Señales de venta de sobrecompra
empieza = '2007-01-01'
termina = '2021-12-31'
cuantos_va = 0
df_global = pd.DataFrame()

window = 21
no_of_std = 2

for ticker in tickers:
    cuantos_va=cuantos_va+1
    df = yf.download(ticker, start=empieza, end=termina).reset_index()
    df['Ticker'] = ticker
    df = calculo_indicadores(df)
    rolling_mean = df['Close'].rolling(window).mean()
    rolling_std = df['Close'].rolling(window).std()
    df['i_BOL_High'] = rolling_mean + (rolling_std * no_of_std)
    df['i_BOL_Low'] = rolling_mean - (rolling_std * no_of_std)
    df['var_3_ruedas'] = df.Close.shift(-3)/df.Close - 1
    df['var_10_ruedas'] = df.Close.shift(-10)/df.Close - 1
    df = df.dropna()
    df_global = df_global.append(df)
```

```python
df_estrategia = df_global[(df_global.High > df_global.i_BOL_High)&(df_global.i_RSI > 80)]
```

```python
print(
len(df_estrategia),
round(len(df_estrategia[df_estrategia.var_3_ruedas < -0.01]) / len(df_estrategia),4),
round(len(df_estrategia[df_estrategia.var_3_ruedas < -0.03]) / len(df_estrategia),4),
round(len(df_estrategia[(df_estrategia.var_3_ruedas > -0.03)&(df_estrategia.var_3_ruedas < 0)]) / len(df_estrategia),4),
round(len(df_estrategia[(df_estrategia.var_3_ruedas > 0)&(df_estrategia.var_3_ruedas < 0.03)]) / len(df_estrategia),4),
round(len(df_estrategia[(df_estrategia.var_3_ruedas > 0.03)]) / len(df_estrategia),4),
round(len(df_estrategia[df_estrategia.var_10_ruedas < -0.03]) / len(df_estrategia),4)
)
```
