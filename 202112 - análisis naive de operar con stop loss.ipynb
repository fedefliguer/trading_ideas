```python
# Todos los activos juntos

import yfinance as yf
import pandas as pd
df_global = pd.DataFrame()

tickers=['GOOGL']

empieza = '2005-01-01'
termina = '2021-11-22'
lags = 30

for ticker in tickers:
    df = yf.download(ticker, start=empieza, end=termina).reset_index()
    df['Ticker'] = ticker
    i = 1
    while i < (lags+1):
        colname = 'c%sb' % (i)                                                  
        df[colname] = round(df['Close'].shift(-i),2)
        i = i + 1
    df_global = df_global.append(df, ignore_index=True)
    
df_global['rendimiento_30d'] = (df_global['c30b'] / df_global['Close']) - 1
df_global['minimo_30d'] = df_global[['c1b','c2b','c3b','c4b','c5b','c6b','c7b','c8b','c9b','c10b','c11b','c12b','c13b','c14b','c15b','c16b','c17b','c18b','c19b','c20b','c21b','c22b','c23b','c24b','c25b','c26b','c27b','c28b','c29b','c30b']].min(axis=1)
df_global['rendimiento_min30d'] = (df_global['minimo_30d'] / df_global['Close']) - 1
df_global_rendimientos = df_global[['Date', 'Ticker', 'rendimiento_30d', 'rendimiento_min30d']].dropna()

import numpy as np
sl = []
results_sl = []

for i in range(1, 1000):
    sl.append(i/1000)
    df_global_rendimientos = df_global[['Date', 'Ticker', 'rendimiento_30d', 'rendimiento_min30d']].dropna()
    df_global_rendimientos['rendimiento_30d'] = np.where(df_global_rendimientos['rendimiento_min30d'] < -i/1000, -i/1000, df_global_rendimientos['rendimiento_30d'])
    results_sl.append(round(df_global_rendimientos.rendimiento_30d.mean(), 4))
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame(dict(
    x = sl,
    y = results_sl
))
fig = px.line(df, x="x", y="y", title="SL y rendimientos") 
fig.show()
```
