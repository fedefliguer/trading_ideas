```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import date, timedelta
import pandas as pd
import numpy as np
import sys

# Fechas
empieza = '2021-01-01'
termina = '2023-03-31'

# Tickers (corto)
tickers = ['AAVE-USD','ADA-USD','ALGO-USD','AMP-USD','ANKR-USD','APE-USD','ATOM-USD','AUDIO-USD','AVAX-USD','AXS-USD','BAND-USD','BAT-USD','BCH-USD','BNT-USD','BSV-USD','BTC-USD','BTT-USD','CAKE-USD','CELO-USD','CHZ-USD','CKB-USD','CRV-USD','DASH-USD','DGB-USD','DOGE-USD','DOT-USD','EGLD-USD','ENJ-USD','EOS-USD','ETC-USD','ETH-USD','FARM-USD','FET-USD','FIL-USD','FLOW-USD','FTM-USD','GLM-USD','GRT-USD','HBAR-USD','ICX-USD','JASMY-USD','KAVA-USD','KEEP-USD','KNC-USD','KSM-USD','LINK-USD','LPT-USD','LRC-USD','LTC-USD','LUNA1-USD','MANA-USD','MATIC-USD','MKR-USD','NEO-USD','OCEAN-USD','OMG-USD','ONT-USD','OXT-USD','PERP-USD','POLY-USD','QNT-USD','QTUM-USD','RAY-USD','REN-USD','SAND-USD','SKL-USD','SOL-USD','SPELL-USD','SRM-USD','STMX-USD','SUSHI-USD','TRAC-USD','TRX-USD','UMA-USD','UNI-USD','VET-USD','VGX-USD','WAVES-USD','XLM-USD','XMR-USD','XTZ-USD','XVG-USD','YFI-USD','YFII-USD','YGG-USD','ZEC-USD','ZEN-USD','ZRX-USD']
df_global = pd.DataFrame()
cuantos_va = 0
for ticker in tickers:
    cuantos_va=cuantos_va+1
    df = yf.download(ticker, start=empieza, end=termina).reset_index()
    df['Ticker'] = ticker
    df = df[df.Volume > 0] # A veces incluye los feriados ARG con Volumen 0, los excluyo
    df_global = df_global.append(df, ignore_index=True)
    if len(df) == 0:
        print("No corrió ", ticker," numero ",cuantos_va," de ", len(tickers))
    else:        
        print("Corrió ", ticker," numero ",cuantos_va," de ", len(tickers))        
```

```python
dias_analisis = 7
df_global['close_diff'] = np.where(df_global.Ticker == df_global.Ticker.shift(-dias_analisis), df_global.Close.shift(-dias_analisis)/df_global.Close - 1, np.nan)
df_global['volume_diff_1db'] = np.where(df_global.Ticker == df_global.Ticker.shift(1), df_global.Volume / df_global.Volume.shift(1), np.nan)
df_global.groupby(pd.qcut(df_global['volume_diff_1db'], 10, labels = False)).agg({'close_diff': ['count', 'mean', 'median']})
```
